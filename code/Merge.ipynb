```python
import pandas as pd
```

```python
# grabbing both csvs to create data frames
deathData = pd.read_csv('../data/parkDataWDeaths.csv')
parkData = pd.read_csv('../data/data.csv')
```

```python
mergedData1 = deathData.merge(parkData, how = 'inner', on = 'Park_Name')
```

```python
mergedData1.to_csv("../data/mergedDeathData.csv")
```

```python
mergedData1.designation.unique()
```

```python
# verfiying all deaths are accounted for 
sum(mergedData1["Death Total"])
```

```python
#  merging data with deaths

mergedData2 = deathData.merge(parkData, how = 'right', on = 'Park_Name')
mergedData2.head()
```

```python
# simplifying data 
mergedData2 = mergedData2[["Park_Name","Death Total","activities"]]
mergedData2.head()
```

```python
# filling NaN fields with int so that it can work with machine learning
x = mergedData2["Death Total"].fillna(0).astype(int)
mergedData2["deaths"] = x
mergedData2["Death Total"] = x
mergedData2
```

```python
# creating binary output for deaths 
mergedData2['deaths'].values[mergedData2['deaths'] > 1] = 1
```
